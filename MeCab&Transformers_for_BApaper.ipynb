```python
##トークナイザーの読み込み

from transformers import BertTokenizer
```

```python
##トークナイザーバージョン指定
tokenizer = BertTokenizer.from_pretrained("cl-tohoku/bert-base-japanese-v3")
```

```python
##例文
tokenizer.tokenize("私の得意科目は数学です")
```

```python
##トークナイザーによるid割り振り(BERTに突っ込める形にする)
tokenizer("私の得意科目は数学です")
```

```python
##複数の文の識別id(token_type_ids)
tokenizer("私の得意科目は数学です。その中でも最も得意なのは代数です。")
```

```python
##トークン列数指定とパディングトークン
tokenizer("私の得意科目は数学です。",max_length=32,padding="max_length")
```

データの文字数を最大32に揃え、不足分は0となるパディングトークンの生成。imput_idsの0はパディングトークンとなる。attention_maskも0になる。

```python
##padding&trancating
tokenizer("私の得意科目は数学です。",max_length=4,trancating="max_length")
```
